```python
import glob
import sys
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
files = "data/"
```

```python
data_files = glob.glob(files+"/2020-02*.csv.gz")
print(data_files)
```

```python
reqCols = ['TweetID', 'CreatedAt', 'CreatedAt_Full','Hashtags']
```

```python
df = pd.concat((pd.read_csv(file, usecols=reqCols) for file in data_files), ignore_index=True)
#data for 21-01-2020
df.head()
```

```python
df['Date'] = df['CreatedAt']
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df['Year'] = df.index.year
df['month'] = df.index.month
df['day'] = df.index.day
```

```python
df.dropna(subset=['Hashtags'], inplace=True)
```

```python
df.assign(Hashtags=df.Hashtags.str.split(",")).explode('Hashtags')
```
