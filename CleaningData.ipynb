```python
# Path
path = "data/cleaned_data"
if not os.path.exists(path):
    os.mkdir(path)
else:
    print("Directory Exists")
```

```python
data_files = glob.glob(files+"/*.csv.gz")
print(data_files)
```

```python
for file in data_files:
    df = pd.read_csv(file)
    #modified dataframe is df2
    df2 = df.assign(Hashtags=df.Hashtags.str.split(",")).explode('Hashtags')

    df2['Hashtags'].str.strip()
    df2.Hashtags = df2.Hashtags.str.lower()
    cleanedFileLoc = "data/cleaned_data/" + file[5:-7] + 'mod.csv.gz'
    #re-writing it back to files 
    df2.to_csv(cleanedFileLoc, 
           index=False, 
           compression="gzip")
```
