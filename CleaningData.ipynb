```python
# Path
path = "data/cleaned_data"
if not os.path.exists(path):
    os.mkdir(path)
else:
    print("Directory Exists")
```

```python
data_files = glob.glob(files+"/*.csv.gz")
print(data_files)
```

```python
for file in data_files:
    df = pd.read_csv(file)
    #modified dataframe is df2
   
    df2 = df.assign(Hashtags=df.Hashtags.str.split(",")).explode('Hashtags')
    # if 'Hashtags' in df2.columns:
    df2.dropna(subset=['Hashtags'], inplace=True)

        
    df2['Hashtags'].str.strip()
    df2.Hashtags = df2.Hashtags.str.lower()

    
    #creating new df column which is called date
    df2['Date'] = df2['CreatedAt']
    #converting it to pandas datetime
    df2['Date'] = pd.to_datetime(df2['Date'])
    df2 = df2.set_index('Date')
    df2['Year'] = df2.index.year
    df2['month'] = df2.index.month
    df2['day'] = df2.index.day

    #creating a time column
    time = df2.CreatedAt_Full.str.split(' ')
    #creating a new column with full date
    df2['date'] = df2['CreatedAt'] + " "+ time.apply(lambda x:x[3])

    cleanedFileLoc = "data/cleaned_data/" + file[5:-7] + 'mod.csv.gz'
    #re-writing it back to files 
    df2.to_csv(cleanedFileLoc, 
           index=False, 
           compression="gzip")
```
